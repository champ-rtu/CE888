```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
from sklearn import cluster, datasets, metrics
import seaborn as sns
```

```python
#Load Country-data.csv
data = pd.read_csv("https://raw.githubusercontent.com/sagihaider/CE888_2021/main/Lab_6/Data/Country-data.csv")
data.head()
```

```python
#Split data for training 
x=data.iloc[:,1:10] 
```

```python
x
```

```python
#use K-mean approach for clustering countried into 3 class 
#1. Underdeverloping  2. developing country  3.developed country
km = cluster.KMeans(n_clusters=3)
km.fit(x)
km.labels_
label = km.labels_
```

```python
data["Label"] = label
```

```python
data['Label'] = data['Label'].map({0:'Under developing', 1:'deverloping country', '2':'developed country'})
```
